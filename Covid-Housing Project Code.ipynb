```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
#Loading data
covid_data = "Resources/covid_report.csv"

covid_df = pd.read_csv(covid_data)

covid_df
```

```python
#Narrowing focus to only counties we want
county_list = ['Marion', 'Hancock', 'Johnson', 'Hamilton', 'Hendricks', 'Boone']
filtered_covid_df = covid_df[covid_df['COUNTY_NAME'].isin(county_list)]
filtered_covid_df

#Dropping Location ID column and resetting index
filtered_covid_df = filtered_covid_df[['GENDER','AGEGRP', 'DATE', 'COVID_COUNT', 'COUNTY_NAME']]
filtered_covid_df

filtered_covid_df.reset_index(drop = True, inplace = True)
filtered_covid_df
```
